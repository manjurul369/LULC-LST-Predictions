```python
def generate_series(mean, std, size, min_val=None, max_val=None, round_to=None):
    values = np.random.normal(loc=mean, scale=std, size=size)
    if min_val is not None:
        values = np.clip(values, min_val, max_val)
    if round_to is not None:
        values = np.round(values, round_to)
    return values
```

```python
new_data = {
    'Point_ID': new_point_ids,
    'Latitude': generate_series(mean=22.8175, std=0.05, size=num_new, min_val=22.65, max_val=22.99, round_to=5),
    'Longitude': generate_series(mean=86.3725, std=0.07, size=num_new, min_val=86.15, max_val=86.55, round_to=5),
    'LST (°C)': generate_series(mean=31.7, std=5, size=num_new, min_val=24, max_val=42, round_to=1),
    'Elevation': generate_series(mean=190, std=25, size=num_new, min_val=140, max_val=240, round_to=0),
    'Slope': generate_series(mean=5.6, std=2, size=num_new, min_val=1, max_val=10, round_to=1),
    'Aspect': generate_series(mean=180, std=100, size=num_new, min_val=0, max_val=360, round_to=0),
    'EVI': generate_series(mean=0.36, std=0.15, size=num_new, min_val=0.05, max_val=0.7, round_to=2),
    'NDVI': generate_series(mean=0.42, std=0.2, size=num_new, min_val=0.1, max_val=0.75, round_to=2),
    'TVI': generate_series(mean=92.25, std=20, size=num_new, min_val=60, max_val=120, round_to=1),
    'NMDI': generate_series(mean=-0.01, std=0.2, size=num_new, min_val=-0.35, max_val=0.4, round_to=2),
    'OSAVI': generate_series(mean=0.45, std=0.2, size=num_new, min_val=0.1, max_val=0.75, round_to=2),
    'SAVI': generate_series(mean=0.50, std=0.2, size=num_new, min_val=0.15, max_val=0.85, round_to=2),
    'BUI': generate_series(mean=0.06, std=0.2, size=num_new, min_val=-0.2, max_val=0.35, round_to=2),
    'EBBI': generate_series(mean=0.04, std=0.2, size=num_new, min_val=-0.2, max_val=0.3, round_to=2),
    'NBI': generate_series(mean=0.0, std=0.2, size=num_new, min_val=-0.25, max_val=0.25, round_to=2),
    'NBAI': generate_series(mean=-0.01, std=0.15, size=num_new, min_val=-0.25, max_val=0.25, round_to=2),
    'UI': generate_series(mean=0.03, std=0.25, size=num_new, min_val=-0.3, max_val=0.45, round_to=2),
    'MNDWI': generate_series(mean=0.11, std=0.3, size=num_new, min_val=-0.4, max_val=0.6, round_to=2),
    'NDWI': generate_series(mean=0.08, std=0.3, size=num_new, min_val=-0.45, max_val=0.65, round_to=2),
}
```
